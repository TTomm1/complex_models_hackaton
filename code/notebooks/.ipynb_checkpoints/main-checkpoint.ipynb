```python
%matplotlib inline  

import numpy as np
import pandas as pd
import sklearn
import sys
import os.path
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
basepath = os.path.dirname(".")
filepath = os.path.abspath(os.path.join(basepath, "..", "data", "train_set.csv"))
print(filepath)
```

```python
traindata=pd.read_csv(filepath)
```
